#Removing Missing Values:

The dataset I will be using is banking data. The dataset appears to contain missing data that needs to be removed.

The following lines of code import the dataset and establishes the column names.

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/Jett987/Machine_Learning/main/Train_removed.csv', header=None)
data.columns = ['id', 'customer_age', 'job_type', 'marital','education', 'default', 'balance', 'housing_loan','personal_loan', 'communication_type','data_of_month','month','last_contact_duration','num_contants_in_campaign','days_since_prev_campaign_contact','num_contacts_prev_campaign','prev_campaign_outcome','term_deposit_subscribed']

print('Total number of Records: %d' % (data.shape[0]))
print('Total number of Features: %d' % (data.shape[1]))
data.head()
```

Next, I will check the dataset for missing values. This dataset contains a significant amount of missing values in the "days_since_prev_campaign_contact" column, and missing values in some of the other columns as well.

```python
import numpy as np

data = data.replace('?',np.NaN)

print('Total Number of Records: %d' % (data.shape[0]))
print('Total Number of Features: %d' % (data.shape[1]))
print('Total Number of missing values:')
for col in data.columns:
    print('\n%s: %d' % (col,data[col].isna().sum()))
```

Next, I will resolve the missing values using median. I will display a small sample of the data before replacing the missing values, and then after replacing the missing values.

```python
data2 = data['customer_age']

print('Before replacing missing values:')
print(data2[273:292])
data2 = data2.fillna(data2.median())
print('\nAfter replacing missing values:')
print(data2[273:292])
```

```python
data2 = data['balance']

print('Before replacing missing values:')
print(data2[1350:1360])
data2 = data2.fillna(data2.median())
print('\nAfter replacing missing values:')
print(data2[1350:1360])
```

```python
data2 = data['last_contact_duration']

print('Before replacing missing values:')
print(data2[25:30])
data2 = data2.fillna(data2.median())
print('\nAfter replacing missing values:')
print(data2[25:30])
```

```python
data2 = data['num_contants_in_campaign']

print('Before replacing missing values:')
print(data2[130:140])
data2 = data2.fillna(data2.median())
print('\nAfter replacing missing values:')
print(data2[130:140])
```

Because the "days_since_prev_campaign_contact" column contains so many missing values, I will drop that feature with the following line of code.

```python
data=data.drop(['days_since_prev_campaign_contact'],axis=1)
data.head()
```

The next line of code will drop rows with the NaN values.

```python
print('Total number of rows before removing missing values: %d' % (data.shape[0]))
data2 = data.dropna()
print('Total number of rows after removing missing values: %d' % (data2.shape[0]))
```

The next line of code will display which features contain a data type other than a numeric data type.

```python
data.dtypes
```

#Checking for Outliers:

Next, I will use box plots to identify any outliers. I will also drop and convert the non-numeric datatypes so that they can be plotted.

When evaluating the boxplot listed above, the outliers appear to be in the $balance$ and $last$ $contact$ $duration$ columns.

```python
Z = (data2-data2.mean())/data2.std()
Z[1:5495]
```

#Checking for Duplicates:

Next, I will now check for duplicated rows.

```python
dups = data.duplicated()
print('Total number of duplicate rows: %d' % (dups.sum()))
```
